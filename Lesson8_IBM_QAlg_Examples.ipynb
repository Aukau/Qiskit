```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
theta = 1 / 9
t = np.arange(0, 300, 1)
y = np.sin((2 * t + 1) * theta)
```

```python
print("Scatter plot")
plt.plot(t, y, "o", color="black")
plt.show()
```

```python
print("Linear interpolation")
plt.plot(t, y)
plt.show()
```

```python
for n in range(1, 20):
    N = 2**n
    theta = np.arcsin(np.sqrt(1 / N))
    t = m.floor(np.pi / (4 * theta))
    p = np.sin((2 * t + 1) * theta) ** 2
    print("%d\t %12.10f" % (N, p))
```

```python
for n in range(2, 20):
    N = 2**n
    old_theta = np.arcsin(np.sqrt(1 / N))
    new_theta = np.arcsin(np.sqrt(4 / N))
    t = m.floor(np.pi / (4 * old_theta))
    p = np.sin((2 * t + 1) * new_theta) ** 2
    print("%d\t %12.10f" % (N, p))
```

```python
s = 7

for n in range(m.ceil(m.log2(s)), m.ceil(m.log2(s)) + 20):
    N = 2**n
    theta = np.arcsin(np.sqrt(s / N))
    t = m.floor(np.pi / (4 * theta))
    p = np.sin((2 * t + 1) * theta) ** 2
    print("%d\t %12.10f" % (N, np.sin((2 * t + 1) * theta) ** 2))
```
